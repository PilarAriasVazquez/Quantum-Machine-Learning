# Quantum Kernel Estimation

In this case, we do not train a model as we did before with the Variational Quantum Classifier, but we use a quantum computer to calculate the similarity matrix, so an SVM can run the rest of the problem.

In a quantum computer we find the Hilbert Space. Thanks to the feature map, we transform classic data x_1 and x_2 to a quantum space using a fixed circuit (feature map) $\phi(x)$.

The quanum kernel is the scalar product (overlap) from those both states:

$$K(x_1, x_2) = |\langle \phi(x_1) | \phi(x_2) \rangle|^2$$


The quantum kernel methos refers to any method that uses quantum computers to estimate a kernel. The goal of the mapping is to make the data categories separable by a hyperplane (the essence of SVM). This scalar product in an infinite dimension space is impossible to calculate for a classic computer, but for a quantum computer it has just to measure the similarity between both physical states.

The main steps are:
1. Calculate the kernel's matrix: calculate similarities between pair of numbers of the dataset using a quantum computer, getting as a result a NxN size matrix.
2. Classical optimization: give to the SVM the matrix
3. Prediction: SVM finds the optimal hyperplane in the quantum space.
